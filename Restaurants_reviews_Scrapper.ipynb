# GETTING SPECIFED RESTAURANTS REVIEWS 

```python
def get_restaurant_reviews(restaurant_id,url,no_of_reviews):
    no_of_reviews=no_of_reviews.replace(",", "")
    no_of_reviews=no_of_reviews[:no_of_reviews.index(' ')]
    Total_reviews = int(no_of_reviews)
    print(Total_reviews)
    reviews_under_one_page=15
    Total_pages_to_scrap_for_reviews=math.floor((Total_reviews/reviews_under_one_page))
    driver.get(url)
    # Retrieve the HTML content of the page after the dynamic content has loaded
    html = driver.page_source
    # Use Beautiful Soup to parse the HTML content
    soup = BeautifulSoup(html, "html.parser") 
    reviews=[]
    restaurant_ids=[]
    
    for item in soup.select(".entry"):
        review=""
        item=list(item)
        review=item[0].text
        if(len(item)>1):
            review+=item[1].text
        reviews.append(review)
        restaurant_ids.append(restaurant_id)
    
    
    for i in range(Total_pages_to_scrap_for_reviews):
        element = driver.find_element(By.CSS_SELECTOR, ".nav.next.ui_button.primary")
        if element:
            try:
                driver.execute_script("arguments[0].click();", element)
            except Exception:
                element = driver.find_element(By.CSS_SELECTOR, ".nav.next.ui_button.primary")
                driver.execute_script("arguments[0].click();", element)
            
            html = driver.page_source
            soup = BeautifulSoup(html, "html.parser")
            for item in soup.select(".entry"):
                review=""
                item=list(item)
                review=item[0].text
                if(len(item)>1):
                    review+=item[1].text
                reviews.append(review)
                restaurant_ids.append(restaurant_id)
        
    return restaurant_ids,reviews

 
```

```python
df=pd.read_csv('restaurants_details - Copy.csv')
pd.options.display.max_colwidth = 1000
```

```python
ids=[]
reviews=[]
for index, row in df.iterrows():
    idslist, reviewslist=get_restaurant_reviews(df['Restaurant ID'][index],df['Restaurant Reviews Link'][index],df['Total Reviews'][index])
    ids.append(idslist)
    reviews.append(reviewslist)
```

```python
reviews_List = []
for sublist in reviews:
    reviews_List.extend(sublist)

Ids_List = []
for sublist in ids:
    Ids_List.extend(sublist)
    
```

```python
reviews_DataFrame=pd.DataFrame(zip(Ids_List,reviews_List))
reviews_DataFrame = reviews_DataFrame.rename(columns={0: 'Restaurant ID', 1: 'Reviews'})
reviews_DataFrame
```

# OPENING THE PROFILE LINK AND GETTING THE FURTHER DETAILS

```python
def get_restaurant_details(url):
    driver.get(url)
    driver.implicitly_wait(10)
    # Retrieve the HTML content of the page after the dynamic content has loaded
    html = driver.page_source
    # Use Beautiful Soup to parse the HTML content
    soup = BeautifulSoup(html, "html.parser")    
    rating=soup.select('.QEQvp .ZDEqb') 
    if(len(rating)>0):
        rating=rating[0].text
    else:
        rating="None"
    
    no_od_reviews=soup.select('.IcelI')
    if(len(no_od_reviews)>0):
        no_od_reviews=no_od_reviews[0].text
    else:
        no_od_reviews="None"
        
    address=soup.select('.yEWoV')
    if(len(address)>0):
        address=address[0].text
    else:
        address="None"
    additonal_address_details=soup.select('.yEWoV.OkcwQ')
    if(len(additonal_address_details)>0):
        additonal_address_details=additonal_address_details[0].text
    else:
        additonal_address_details="none"
    
    website=soup.select('.IdiaP.Me.sNsFa span a')
    if(len(website)>0):
        website = str(website[0].get("href"))
    else:
        website="None"
        
    phone=list(soup.select('.f .IdiaP.Me a'))
    
    if(len(phone)>1):
        phone=str(phone[1].get("href"))
    elif(len(phone)>2):
        phone=str(phone[2].get("href"))
    else:
        phone="None"
        
    if("mailto" in phone):
        phone=phone[7:-10]
    information=[rating,no_od_reviews,address,additonal_address_details,website,phone]
    return information
```
